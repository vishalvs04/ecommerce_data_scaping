# LAPTOP DATA SCRAPING FROM COMP

```python
laptop_url='https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
laptop_req=requests.get(laptop_url)
laptop=BeautifulSoup(laptop_req.text,'html.parser')
```

```python
laptop
```

```python
laptop_title=laptop.find_all('a',class_='title')
```

```python
laptop_description=laptop.find_all('p',class_='description')
```

```python
laptop_price=laptop.find_all('h4',class_='pull-right price')
laptop_reviews=laptop.find_all('p',class_='pull-right')
```

```python
final_laptop_title=[]
final_laptop_description=[]
final_laptop_price=[]
final_laptop_reviews=[]
for i in range(len(laptop_title)):
    final_laptop_title.append((laptop_title[i].string).strip())
    final_laptop_description.append((laptop_description[i].string).strip())
    final_laptop_price.append((laptop_price[i].string).strip())
    final_laptop_reviews.append((laptop_reviews[i].string).strip())
```

```python
final_deets=[]
for i in range(len(final_laptop_title)):
    final_deets.append([final_laptop_title[i],final_laptop_description[i],final_laptop_price[i],final_laptop_reviews[i]])
```

```python
final_deets
```

# PHONE DATA SCRAPING FROM COMP 

```python
phones_url='https://webscraper.io/test-sites/e-commerce/allinone/phones'
phones_req=requests.get(phones_url)
phones=BeautifulSoup(phones_req.text,'html.parser')
```

```python
print(phones.prettify())
```

```python
touch_phones_url='https://webscraper.io/test-sites/e-commerce/allinone/phones/touch'
touch_phones_req=requests.get(touch_phones_url).text
touch_phones=BeautifulSoup(touch_phones_req,'html.parser')
```

```python
phone_name=touch_phones.find_all('a',class_='title')
phone_price=touch_phones.find_all('h4', class_='pull-right price')
phone_des=touch_phones.find_all('p', class_='description')
phone_rating=touch_phones.find_all('p',class_='pull-right')
```

```python
final_phone_names=[]
final_phone_price=[]
final_phone_description=[]
final_phone_reviews=[]
for i in range(len(phone_name)):
    final_phone_names.append((phone_name[i].string).strip())
    final_phone_description.append((phone_des[i].string).strip())
    final_phone_price.append((phone_price[i].string).strip())
    final_phone_reviews.append((phone_rating[i].string).strip())
```

```python
phone_deets=[]
for i in range(len(final_phone_names)):
    phone_deets.append([final_phone_names[i],final_phone_description[i],final_phone_price[i],final_phone_reviews[i]])
```

```python
phone_deets
```

# TABLETS DATA SCRAPING FROM COMP

```python
tablets_url='https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
tablets_req=requests.get(tablets_url)
tablets=BeautifulSoup(tablets_req.text,'html.parser')
```

```python
tablets
```

```python
tablets_name=tablets.find_all('a',class_='title')
tablets_reviews=tablets.find_all('p',class_='pull-right')
tablets_price=tablets.find_all('h4', class_='pull-right price')
tablets_description=tablets.find_all('p',class_='description')
```

```python
final_tablets_name=[]
final_tablets_price=[]
final_tablets_description=[]
final_tablets_reviews=[]
for i in range(len(tablets_name)):
    final_tablets_name.append((tablets_name[i].string).strip())
    final_tablets_description.append((tablets_description[i].string).strip())
    final_tablets_price.append((tablets_price[i].string).strip())
    final_tablets_reviews.append((tablets_reviews[i].string).strip())
```

```python
tablets_data=[]
for i in range(len(tablets_name)):
    tablets_data.append([final_tablets_name[i],final_tablets_price[i],final_tablets_description[i],final_tablets_reviews[i]])
```

```python
tablets_data
```

# CREATING CSV FOR EACH CATEGORY OF PRODUCTS

```python
laptop_file_writer.writerows(final_deets)
```
